```python
def process_covid_ts(input_filename, output_filename):
    df = pd.read_csv(input_filename)
    df["Location"] = df["Country/Region"] + " - " + df["Province/State"]
    df.loc[df["Location"].isnull(),"Location"] = df["Country/Region"]
    df.to_csv(output_filename, sep=',', encoding='utf-8')
```

```python
process_covid_ts("time_series_covid19_confirmed_global.csv", "processed_covid19_confirmed_global.csv")
process_covid_ts("time_series_covid19_deaths_global.csv", "processed_covid19_deaths_global.csv")
process_covid_ts("time_series_covid19_recovered_global.csv", "processed_covid19_recovered_global.csv")
```

```python
def process_covid_active_ts(input_filename):
    df = pd.read_csv(input_filename)
    df = df.groupby('Country/Region').sum()
    return df
```

```python
confirmed = process_covid_active_ts("time_series_covid19_confirmed_global.csv")
death = process_covid_active_ts("time_series_covid19_deaths_global.csv")
recovered = process_covid_active_ts("time_series_covid19_recovered_global.csv")
```

```python
active = confirmed - death - recovered
```

```python
active.to_csv("active_cases_ts.csv", sep=",", encoding="utf-8")
```
